```python
from pathlib import Path
import pandas as pd
import numpy as np
```

```python
BASE_DIR = Path('../../../')

CSV_DIR = BASE_DIR / 'data' / 'iemocap_4way_data'
TRAIN_CSV_PATH = CSV_DIR / 'train_4way_with_minus_one.csv'
VAL_CSV_PATH = CSV_DIR / 'val_4way_with_minus_one.csv'
TEST_CSV_PATH = CSV_DIR / 'test_4way_with_minus_one.csv'

EMBEDDING_DIR = BASE_DIR / 'data' / 'embeddings' / '4way_opensmile' / 'eGeMAPSv02'
TRAIN_EMBEDDING_PATH = EMBEDDING_DIR / 'train_filtered.npy'
VAL_EMBEDDING_PATH = EMBEDDING_DIR / 'val_filtered.npy'
TEST_EMBEDDING_PATH = EMBEDDING_DIR / 'test_filtered.npy'
```

### Number of samples in CSV files vs. Embedding files

```python
def count_csv_rows(filepath):
    df = pd.read_csv(filepath)
    return len(df)

def count_npy_rows(filepath):
    arr = np.load(filepath, allow_pickle=True)
    return arr.shape[0]
```

```python
print(f"Training data rows: {count_csv_rows(TRAIN_CSV_PATH)}")
print(f"Validation data rows: {count_csv_rows(VAL_CSV_PATH)}")
print(f"Testing data rows: {count_csv_rows(TEST_CSV_PATH)}")
```

```python
print(f"Training embedding rows: {count_npy_rows(TRAIN_EMBEDDING_PATH)}")
print(f"Validation embedding rows: {count_npy_rows(VAL_EMBEDDING_PATH)}")
print(f"Testing embedding rows: {count_npy_rows(TEST_EMBEDDING_PATH)}")
```
